```python
import pandas as pd
```

```python
# read the data

df = pd.read_csv("amazon_india_2015_2025_merged_cleaned.csv")
df['order_date'] = pd.to_datetime(df['order_date'])
```

```python
df_products = pd.read_csv("amazon_india_products_catalog.csv")
```

```python
# save transactions csv
transactions = df[[
    'transaction_id',
    'customer_id',
    'product_id',
    'order_date',
    'order_year',
    'original_price_inr',
    'discount_percent',
    'discounted_price_inr',
    'quantity',
    'subtotal_inr',
    'delivery_charges',
    'final_amount_inr',
    'payment_method',
    'delivery_type',
    'delivery_days',
    'is_festival_sale',
    'festival_name',
    'customer_rating',
    'return_status'
    ]]

transactions.to_csv("transactions.csv", index=False)
```

```python
# save products csv

products = df_products[[
    'product_id',
    'product_name',
    'category',
    'subcategory',
    'brand',
    'base_price_2015',
    'rating',
    'is_prime_eligible',
    'launch_year',
    'model',
    'weight_kg'
]]

products.to_csv("products.csv", index=False)
```

```python
# save customers csv

customers = df[[
    'customer_id',
    'customer_city',
    'customer_state',
    'customer_tier',
    'customer_spending_tier',
    'customer_age_group',
    'is_prime_member'
]]
customers = customers.drop_duplicates(subset=['customer_id']).reset_index(drop=True)
customers.to_csv("customers.csv", index=False)
```

```python
# save time dimensions

time_dimension = df[[
    'order_date',
    'order_month',
    'order_quarter',
    'order_year'
]]
time_dimension = time_dimension.drop_duplicates(subset=['order_date']).reset_index(drop=True)
time_dimension.to_csv("time_dimension.csv", index=False)
```
